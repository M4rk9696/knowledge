```python
import casacore.tables as tbl

ms = "1934-638_0.ms"
```

```python
obs = tbl.table("%s/OBSERVATION" % (ms))

obs_data = tbl.taql("select * from $obs")

obs.close()

obs_data
```

```python
# Csimulator.spws.TestBand                        = [48, 1319.129.0MHz, 1.0MHz, "XX YY XY YX"]
# This is simulated as 48 channels, starting from 1319.129.0MHz in increments of 1.0MHz, final bandwidth is 48MHz

# Spectral window of the observation
spw = tbl.table("%s/SPECTRAL_WINDOW" % (ms), ack=False)

content = tbl.taql("select CHAN_FREQ, NUM_CHAN, TOTAL_BANDWIDTH from $spw")

spw.close()

content
```

```python
# Antenna
antenna = tbl.table("%s/ANTENNA" % (ms), ack=False)

# Number of antenna
nAnt = antenna.nrows()
# Number of baselines
nBase = nAnt * (nAnt + 1) // 2

antenna.close()

print(nAnt, nBase)
```

```python
# Data
pt = tbl.table("%s/POINTING" % (ms), ack=False)

content = tbl.taql("select * from $pt limit 10")

pt.close()

content
```
