## Power Spectral Density

```python
import numpy as np
from scipy import signal, misc
import matplotlib.pyplot as plt

fs = 10e3
N = 1e5
amp = 2*np.sqrt(2)
freq = 1270.0
time = np.arange(N) / fs
```

## Power Spectral Density of noise

The power spectral density of noise is a constant


```python
noise_power = 0.001 * fs / 2

noise = np.random.normal(scale=np.sqrt(noise_power), size=time.shape)

f, Pwelch_spec = signal.welch(noise, fs, scaling='spectrum')

plt.semilogy(f, Pwelch_spec)
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD')
plt.grid()
plt.show()
```

## Power Spectral Density of a function without noise

```python
x = amp*np.sin(2*np.pi*freq*time)

f, Pwelch_spec = signal.welch(x, fs, scaling='spectrum')

plt.semilogy(f, Pwelch_spec)
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD')
plt.grid()
plt.show()
```

## Power Spectral Density of a function with noise

```python
sgnl = x + noise

f, Pwelch_spec = signal.welch(sgnl, fs, scaling='spectrum')

plt.semilogy(f, Pwelch_spec)
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD')
plt.grid()
plt.show()
```
